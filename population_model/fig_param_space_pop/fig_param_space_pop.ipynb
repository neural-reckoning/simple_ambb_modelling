# Figure parameter space population model

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from brian2 import *
from scipy.stats import circstd,circmean
from scipy.ndimage.filters import gaussian_filter
from collections import *
from matplotlib.pyplot import *
```

```python
path_to_files='./param_space_pop_data_files/'
tau_b=np.load(path_to_files+'tau_b.npy')
sigma_k=np.load(path_to_files+'sigma_k.npy')
gamma=np.load(path_to_files+'gamma.npy')
k=np.load(path_to_files+'k.npy')
m=np.load(path_to_files+'m.npy')

dict_param=OrderedDict([('tau_b',tau_b),('sigma_k',sigma_k),('gamma',gamma),('k',k),('m',m)])
latex_param_names=OrderedDict([('tau_b',r'$\tau_{b}$'),('sigma_k',r'$\sigma_{k}$'),('gamma',r'$\gamma$'),('k',r'$k$'),('m',r'$m$')])
```

## Preprocessing, averaging over start IPD

```python
def mean_over_phi_start(filename_res,carrier,path):
    res=load(filename_res)
    nbr_param_sets=int(res.size/(res.shape[0]*res.shape[6]))
    res_flat=res.reshape((res.shape[0],nbr_param_sets,res.shape[6]))
    res_flat_average=np.zeros((res.shape[0],nbr_param_sets))
    sd_flat=np.zeros((res.shape[0],nbr_param_sets))
    for i in range(nbr_param_sets):
        res_flat_average[:,i]=circmean(res_flat[:,i,:],high=360,axis=1)
        sd_flat[:,i]=circstd(res_flat[:,i,:],high=360,axis=1)
    res_average=res_flat_average.reshape((res.shape[0],res.shape[1],res.shape[2],res.shape[3],res.shape[4],res.shape[5]))
    sd_average=sd_flat.reshape((res.shape[0],res.shape[1],res.shape[2],res.shape[3],res.shape[4],res.shape[5]))
    save(path+'phase_'+str(carrier)+'.npy',res_average)
    save(path+'sd_'+str(carrier)+'.npy',sd_average)
```

## Computing max error

```python
def max_error(filename_phase,carrier,path):
    if carrier==200:
        phase_data=np.array([180,180,180,180,180])
    else:
        phase_data=np.array([37, 40, 62, 83, 115])
    phase=load(filename_phase)
    nbr_param_sets=int(phase.size/phase.shape[0])
    phase_flat=phase.reshape((phase.shape[0],nbr_param_sets))
    
    max_error_flat=np.zeros(nbr_param_sets)
    cp_nan=0
    for i in range(nbr_param_sets) :
        max_error_flat[i]=amax(abs(phase_flat[:,i]-phase_data))
    max_error=max_error_flat.reshape((1,phase.shape[1],phase.shape[2],phase.shape[3],phase.shape[4],phase.shape[5]))

    save(path+'max_error_'+str(carrier)+'.npy',max_error)
```

```python
def preproc(filename_res_500,path):
    mean_over_phi_start(path+filename_res_500,500,path)
    max_error(path+'phase_500.npy',500,path)
```

## Extracting best solutions 

```python
def best_sols(filename_max_err,dict_param,path,max_error_thresh=30):
    max_err_arr=np.squeeze(np.load(path+filename_max_err))
    num_fm=5
    num_param=len(dict_param)
    b_sols=np.argwhere(max_err_arr<=max_error_thresh)
    b_values=np.zeros((b_sols.shape[0],num_param))
    for k in range(num_param):
        b_values[:,k]=dict_param[list(dict_param.keys())[k]][b_sols[:,k]]
    return b_values,b_sols 
```

## Stats

```python
def isincreasing(L):
    return all(x<=y for x, y in zip(L, L[1:]))
```

```python
def monotony(filename):
    res=load(filename)
    nbr_param=int(1.*res.size/res.shape[0])

    res_flat=res.reshape(res.shape[0],res.shape[1]*res.shape[2]*res.shape[3]*res.shape[4]*res.shape[5])
    cp=0
    for i in range(nbr_param):
        if sum(isincreasing(res_flat[:,i]))==4 and np.all(res_flat[:,i]):
            cp+=1
    pc=100.0*cp/(nbr_param)
    return pc
    
```

```python
def phase_location(filename_phase):
    res=load(filename_phase)
    data_500=array([37, 40, 62, 83, 115])
    nbr_param=int(1.*res.size/res.shape[0])
    res_flat=res.reshape(res.shape[0],int(res.size/res.shape[0]))
    cp_up=0
    cp_down=0
    cp_mid=0
    for i in range(nbr_param):
        if sum(res_flat[:,i]>data_500)==5:
            cp_up+=1
        elif sum(res_flat[:,i]>data_500)==0:
            cp_down+=1
        else:
            cp_mid+=1
    pc_up=100.*cp_up/nbr_param
    pc_down=100.*cp_down/nbr_param
    pc_mid=100.*cp_mid/nbr_param
    return pc_up,pc_down,pc_mid
```

## Plotting phase/fm curves

```python
def phase_curve_plots(filename_phase,filename_sd,filename_max_err,path):
    gs=GridSpec(27,20)
    gs.update(wspace=1.5, hspace=1.5)
    ftsize=14
    transp=0.1
    color='#ff7f0e'
    phase=np.load(path+filename_phase)
    pc=monotony(path+filename_phase)
    pc_up,pc_down,pc_mid=phase_location(path+filename_phase)
    print('Percentage of the phases above the experimental data: ' + str(pc_up) + '%')
    print('Percentage of the phases below the experimental data: ' + str(pc_down)+ '%')
    print('Percentage of the phases not above nor below the experimental data: ' + str(pc_mid) + '%')
    print('Percentage of monotonically increasing phases: ' + str(pc)+'%')
    sd=np.load(path_to_files+filename_sd)

    b_values,b_sols=best_sols(filename_max_err,dict_param,path,360)
    b_values_best,b_sols_best=best_sols(filename_max_err,dict_param,path,amin(np.load(path+filename_max_err)))
    seed(34032483)
    rand_ind=np.random.choice(range(b_sols.shape[0]), int(b_sols.shape[0]/30.), replace=False)
    fm=[4,8,16,32,64]
    exp_data=[37, 40, 62, 83, 115]
    std_exp_data=[46, 29, 29, 31,37]
    subplot(gs[0:8,13:20])
    for i in range(b_sols.shape[0]):
        if i in rand_ind: #Plots of the 1/50 phase/fm curves
            indices=(slice(None),int(b_sols[i,0]),int(b_sols[i,1]),int(b_sols[i,2]),int(b_sols[i,3]),int(b_sols[i,4]))
            if i==0:
                plot(fm,phase[indices],lw=0.5,alpha=0.2,label='Model (all)')

            else:
                plot(fm,phase[indices],lw=0.5,color=(0.4, 0.7, 0.4, transp))

    indices_best=(slice(None),int(b_sols_best[0,0]),int(b_sols_best[0,1]),int(b_sols_best[0,2]),int(b_sols_best[0,3]),int(b_sols_best[0,4]))
    errorbar(fm, exp_data,std_exp_data,lw=2.5, fmt='--r', label='Data')
    plot(fm,phase[indices_best],lw=2.5,marker='o',label='Model (best)',color="k")
    xlabel('Modulation frequency (Hz)',fontsize=ftsize)
    ylabel('Extracted IPD (deg)',fontsize=ftsize)
    xticks([4,8,16,32,64],fontsize=ftsize)
    yticks([0,90,180,270,360],fontsize=ftsize)
    ylim([0,360])
    xlim([4,64])
    grid(True)
    legend(fontsize=ftsize)
```

## Plot parameter histograms

```python
def param_hist(filename_max_err,dict_param,latex_param_names,path,max_thresh=30):
    max_err=load(path+filename_max_err)
    gs=GridSpec(27,20)
    gs.update(wspace=1.5, hspace=1.5)
    b_values,b_sols=best_sols(filename_max_err,dict_param,path,max_thresh)
    perc=100.*b_sols.shape[0]/max_err.size
    print('% of the max errors below 30 deg: ' + str(perc) + '%')
    print('Minimum error is:'+str(amin(max_err)) + ' deg')
    
    mid_param=[0.25,0.05,0.005,0,0.05]
    initial_val=[0.125,0.05,0,0,0]
    num_param=5
    color='#ff7f0e'
    for h in range(num_param):
        if h==3:
            unique, counts = np.unique(b_values[:,h], return_counts=True)    
            ax=subplot(gs[23:27,4*h:4*(h+1)])
            title(latex_param_names[list(latex_param_names.keys())[h]],fontsize=20)
            bar([1.6,3.6,7.6],[counts[0],counts[1],counts[2]],lw=0,alpha=1,color='lightgray')
            plot([np.mean(b_values[:,h]),np.mean(b_values[:,h])],[0,max(counts)],ls='--',lw=4,label='Mean',alpha=1)
            xticks(dict_param[list(dict_param.keys())[h]]-0.5,fontsize=14)
            ax.set_xticklabels([0,4,8])
            yticks([],[])
        else:   
            subplot(gs[23:27,4*h:4*(h+1)])
            title(latex_param_names[list(latex_param_names.keys())[h]],fontsize=20)     
            y, x, _=hist(b_values[:,h],bins=np.insert(dict_param[list(dict_param.keys())[h]]+mid_param[h],0,initial_val[h]),alpha=1,lw=0,color='lightgray')
            plot([np.mean(b_values[:,h]),np.mean(b_values[:,h])],[0,max(y)],ls='--',lw=4,label='Mean',alpha=1)
            xticks([dict_param[list(dict_param.keys())[h]][0],dict_param[list(dict_param.keys())[h]][int(size(dict_param[list(dict_param.keys())[h]])/2)],dict_param[list(dict_param.keys())[h]][-1]],fontsize=14)
            yticks([],[])
```

## Model error plotted as a function of different pairs of parameters

```python
def bfits_plots(filename_max_err,dict_param,path):
    
    gs=GridSpec(27,20)
    gs.update(wspace=1.6, hspace=1.6)
    max_err=np.squeeze(np.load(path+filename_max_err))
    y_axis=[2,3,1,4]
    x_axis=[0,2,3,1]
  
    ax=[]
    ftsize=15
    for i in range(4):
        for j in range(i+1):
            b_fit=np.zeros((dict_param[list(dict_param.keys())[y_axis[i]]].shape[0],dict_param[list(dict_param.keys())[x_axis[j]]].shape[0]))
            for k in range(dict_param[list(dict_param.keys())[y_axis[i]]].shape[0]):
                for l in range(dict_param[list(dict_param.keys())[x_axis[j]]].shape[0]):
                    indices=[slice(None),slice(None),slice(None),slice(None),slice(None)]
                    indices[y_axis[i]]=k
                    indices[x_axis[j]]=l
                    b_fit[k,l]=np.nanmin(max_err[tuple(indices)])
            ax.append(subplot(gs[5*i:5*(i+1),5*j:5*(j+1)]))
            x_extent_min=dict_param[list(dict_param.keys())[x_axis[j]]][0]
            x_extent_max=dict_param[list(dict_param.keys())[x_axis[j]]][-1]
            y_extent_min=dict_param[list(dict_param.keys())[y_axis[i]]][0]
            y_extent_max=dict_param[list(dict_param.keys())[y_axis[i]]][-1]                                              
            error_blur = gaussian_filter(b_fit, 1, mode='nearest')
            cs = contour(error_blur, origin='lower', aspect='auto',levels=[15, 30, 45],linewidths=2.5, colors='w',extent=[x_extent_min,x_extent_max,y_extent_min,y_extent_max])
            for level, linecol in zip(cs.levels, cs.collections):
                if level==15:
                    linecol.set_linestyle('solid')
                if level==30:
                    linecol.set_linestyle('dashed')
                if level==45:
                    linecol.set_linestyle('dotted')
    
            #clabel(cs, colors='w', inline=True,fmt='%d')
            im=imshow(b_fit,origin='lower left',interpolation='nearest', aspect='auto',cmap='viridis',vmin=0,vmax=180,extent=[x_extent_min,x_extent_max,y_extent_min,y_extent_max])
            #if i==0 and j==0:
            #    text(0,0.18,'A',fontsize=28) 
            if 5*j==0:
                ylabel(list(latex_param_names.values())[y_axis[i]],fontsize=ftsize)
                yticks([y_extent_min,y_extent_max],fontsize=ftsize)
            else:
                yticks([])
            if 5*(i+1)==20:
                xlabel(list(latex_param_names.values())[x_axis[j]],fontsize=ftsize)
                xticks([x_extent_min,x_extent_max],fontsize=ftsize)
            else:
                xticks([])
    return im
   
    
```

```python
def figure_param_space_pop(filename_phase,filename_sd,filename_max_err,dict_param,path):
    preproc('ipd_500.npy',path)
    fig=figure(figsize=(15,18))
    im=bfits_plots(filename_max_err,dict_param,path)
    cbar_ax = fig.add_axes([0.125, 0.2725, 0.775, 0.012])
    colorbar(im, cbar_ax,orientation='horizontal')
    cbar_ax.tick_params(labelsize=14)
    ###################################################################################

    param_hist(filename_max_err,dict_param,latex_param_names,path,max_thresh=30)
    
    ####################################################################################
    
    phase_curve_plots(filename_phase,filename_sd,filename_max_err,path)
    
    ####################################################################################
    #hist_production(filename_max_err)
    
    text(-140,400,'A',fontsize=25)
    text( -15,400,'B',fontsize=25)
    text( -140,-750,'C',fontsize=25)
    text(-38,-645,'Max error (deg)',fontsize=15)
    
    #fig.savefig('fig_param_space.pdf',bbox_inches="tight")
```
